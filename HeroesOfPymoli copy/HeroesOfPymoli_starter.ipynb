```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
unique_player_counts = purchase_data['SN'].unique()
Total_Players =len(unique_player_counts)
pd.DataFrame({'Total Players': [Total_Players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item_counts = purchase_data['Item ID'].unique()
Total_Unique_Items =len(unique_item_counts)

Average_Price = round(purchase_data['Price'].mean(),2)
Number_purchases = purchase_data['Item Name'].count()
Total_Revenue = purchase_data['Price'].sum()

df_2= pd.DataFrame({'Number of Unique Items': [Total_Unique_Items],'Average Price': [Average_Price],'Number of Purchases': [Number_purchases],'Total Revenue': [Total_Revenue]})
df_2['Average Price'] = '$' + df_2['Average Price'].astype(str)
df_2['Total Revenue'] = '$' + df_2['Total Revenue'].astype(str)
df_2
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Players = purchase_data['SN'].nunique()
grouped_gender = purchase_data.groupby(['Gender']).nunique()
grouped_SN = grouped_gender['SN']
grouped_SN =pd.DataFrame(grouped_SN).sort_values("SN",ascending=False)
grouped_SN["Percentage of Players"]=(round(grouped_SN['SN']/Players *100,2)).astype(str) +'%'
grouped_SN = grouped_SN.rename(columns={'SN': 'Total_Count'})
grouped_SN
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins= [0,9,14,19,24,29,34,39,50]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
purchase_data["Ages"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
purchase_grouped =purchase_data.groupby('Ages')

grouped_age_3 = purchase_data.groupby(['Ages', 'SN'])
grouped_total_players = pd.DataFrame(round(grouped_age_3['Price'].sum(),2))
grouped_total_players= grouped_total_players.reset_index()
grouped_total_players = grouped_total_players.groupby(['Ages'])

Purchase_Count = purchase_grouped['Purchase ID'].nunique()
avg_purchase_price ='$' + (round(purchase_grouped['Price'].mean(),2)).astype(str)
total_purchase ='$' +(round(purchase_grouped['Price'].sum(),2)).astype(str)
avg_total_purchase_price = '$'+ (round(grouped_total_players['Price'].mean(),2)).astype(str)



summary_table = pd.DataFrame({"Purchase Count": Purchase_Count, "Average Purchase Price": avg_purchase_price, "Total Purchase Value":total_purchase,"Average Total Purchase per Person": avg_total_purchase_price})
summary_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_grouped =purchase_data.groupby('SN')

Purchase_Count = purchase_grouped['Price'].count()
avg_purchase_price ='$' + (round(purchase_grouped['Price'].mean(),2)).astype(str)
total_purchase =round(purchase_grouped['Price'].sum(),2)

summary_table = pd.DataFrame({"Purchase Count": Purchase_Count, "Average Purchase Price": avg_purchase_price, "Total Purchase Value":total_purchase})
summary_table = summary_table.sort_values("Total Purchase Value",ascending=False)
summary_table.loc[:, "Total Purchase Value"] ='$' + summary_table["Total Purchase Value"].astype("str")
summary_table
```
